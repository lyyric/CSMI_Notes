## Histogrammes



Considérons maintenant un échantillon, c-à-d un ensemble de nombres. Dans la plupart des cas les échantillons sont construits:

* soit à partir de simulations successives de v.a (ex: v.a gaussiennes)
* soit à partir d'observations (ex: tailles des gens dans la rue)

Dresser l'histogramme d'un échantillon consiste à créer une partition de sous-intervalles, puis d'afficher des bâtons qui ont pour base ces sous-intervalles, et comme hauteur le nombre d'éléments de l'échantillon contenus dans chaque sous-intervalle.




### Découpage automatique

Ci-dessous, `bins=n` signifie que l'on découpe l'intervalle `[min(X),max(X)]` en ¡`n` sous-intervalles.

```python
"""l'échantillon à observer"""
X=np.random.normal(0,1,size=1000)
plt.hist(X,bins=15,edgecolor="k");
```

### Récupérer les valeurs numériques

Les fonctions `plt.hist()` et `np.histogram()` (idem mais sans affichage) permettent aussi de récupérer des valeurs numériques.

```python
a=plt.hist(X,bins=5,edgecolor="k")
print("hauteur des bâtons\n",a[0])
print("découpage\n",a[1])
b=np.histogram(X,bins=5)
print("hauteur des bâtons\n",b[0])
print("découpage\n",b[1])
```

###  `np.unique`

`np.unique(x)` renvoie les valeurs de `x` sans doublons. Au passage, elle peut :
* renvoyer le rang de la première occurence de chaque valeur avec l'option `return_index=True`,
* renvoyer le rang de chaque valeur de `x` dans la liste des valeurs uniques, de manière à pouvoir reconstruire `x`, avec l'option`return_inverse=True`,
* compter et renvoyer le nombre d'occurences de chaque valeur avec l'option `return_counts=True`.

Exemple :

```python
x = np.array([4,4,6,6,4,6,8])
val, index, inverse, count = np.unique(x, return_counts=True, return_index=True, return_inverse=True)

print("x       ", x)
print("val     ", val)
print("index   ", index)
print("inverse ", inverse)
print("count   ", count)
```

### Question de math

Montrons qu'une telle loi géométrique, c'est quasiment une loi exponentielle.


La fonction de répartition de la loi géométrique est
$$
k \to 1 - (1-p)^k = 1 - e^{- k |\ln(1-p)|  }
$$
La fonction de répartition de la loi exponentielle est:
$$
x \to 1 - e^{-x \lambda }
$$
Ainsi, en choisissant $\lambda= |\ln(1-p)|$:

* les deux fonctions de répartition coïncident sur les entiers ;
* les probabilités $\mathbf P[k< X \leq k+1]$  coïncident pour les deux lois ;
* la densité de la loi exponentielle est proche de la densité discrète de la loi géométrique ;
* l'histogramme d'un échantillon géométrique sera proche de la densité de la loi exponentielle.

On peut vérifier expérimentalement le dernier point, en superposant un histogramme d'une loi géométrique avec la densité de la loi exponentielle:

### La gaussienne

L'option importante pour que l'histogramme se superpose à une densité c'est `density=True`

```python
nbSimu=1000
Simu=np.random.normal(size=nbSimu)

"""formule à emmener partout avec soi"""
def gaussian_density(x):
    return 1/(np.sqrt(2*np.pi))*np.exp(-0.5*x**2)

plt.hist(Simu,bins=30,density=True,edgecolor="k");

x=np.linspace(-3,3,200)
plt.plot(x, gaussian_density(x));
```

### La densité de la log normale.

#### →♡♡♡



Calculons la densité de la loi log-normale. Soit $\phi$ une fonction test et $X$ une v.a de loi normale centrée réduite. On a
$$
     \mathbf   E [\phi( \exp(X) )] =  \frac{1}{\sqrt{2\pi}}  \int \ \color{red}{\square \square \square} \  e^{- \frac {1}{2} x^2 }\ dx.
$$
Après changement de variable $e^x \to  y$ on obtient
$$
     \mathbf   E [\phi( \exp(X) )] =  \frac{1}{\sqrt{2\pi}}  \int \color{red}{\square \square \square},
$$
ººuntil:
donc la densité de $\exp(X)$ est $y \mapsto \frac{1}{\sqrt{2\pi}}\cdot e^{-\frac{1}{2}ln^2(y)}\cdot \frac{1}{y}$. C'est la fonction qui a été tracée ci-dessus, superposée aux histogrammes.

```python
size = 10000
Y = np.random.lognormal(size=size)
bins = np.linspace(0,10,80)
plt.hist(Y, bins=bins, edgecolor="k",label=["Y"],density=True)



y = np.linspace(0.001,10,1000)
plt.plot(y, 1/np.sqrt(2*np.pi)*np.exp(-0.5*np.log(y)**2)/y);
```

## Estimons une densité

La technique la plus répandue pour estimer une densité est de convoluer un noyau (gaussien le plus souvent) avec la mesure empirique, ce qui signifie qu'on somme des petites gaussiennes autour de chaque point observé. La largeur (=écart-type=`bande-width`) des noyaux est très importante:
* trop large: la densité est trop lissée,
* trop serré: la densité est trop irrégulière.

```python
X=np.random.beta(a=0.5,b=2,size=15)
fig,ax=plt.subplots(1,1,figsize=(10,3))
ax.plot(X,np.zeros_like(X),'.')

x=np.linspace(np.min(X),np.max(X),200)

kernel=stats.gaussian_kde(X,bw_method=0.1)
y=kernel(x)
ax.plot(x,y)

kernel2=stats.gaussian_kde(X,bw_method=0.2)
y2=kernel2(x)
ax.plot(x,y2);
```

Par défaut, `gaussian_kde()` calcule `bw_method` avec la technique de Scott, qui consiste à prendre  
$$
n^{ \frac{-1}{d+4} },
$$
avec $n$ le nombre de donnée, et $d$ la dimension (ici 1, plus loin 2).

### Des données trajectorielles $\triangleright$

```python
def makeTrajectories(T,nbSimu):
    """Pour chaque instant de chaque simulation on tire au hasard un "numéro de pas" (0, 1, 2 ou 3)"""
    pre_pas=np.random.randint(0, 4, size=[nbSimu,T])
    pas=np.zeros(shape=[nbSimu,T, 2])
    """On remplace chaque "numéro de pas" par le mouvement correspondant sous la forme [x,y] """
    pas[pre_pas==0]=[2,1]
    pas[pre_pas==1]=[-1,0]
    pas[pre_pas==2]=[0,1]
    pas[pre_pas==3]=[0,-1]
    """On fait les sommes cumulées selon l'axe du temps pour avoir la position à chaque instant"""
    return np.cumsum(pas,axis=1)


def test():
    trajectories=makeTrajectories(T=7,nbSimu=3)
    print(trajectories.shape)
    print(trajectories)
```

Dans le code ci-dessous, remplacez `T=100` par `T=200`. Que représente ce paramètre `T` ?

```python
trajectories=makeTrajectories(T=100,nbSimu=20)
for tr in trajectories:
    plt.plot(tr[:,0],tr[:,1])
```

On observe que les trajectoires ne peuvent arriver que dans "une case sur deux". Plus précisément, toutes les trajectoires arrivent dans le sous-ensemble $$
P = \{ (x,y)\in\mathbb{Z}\times\mathbb{Z}\ ; x+y \text{ est pair}\},
$$
et aucune n'arrive dans son complémentaire
$$
I = \{ (x,y)\in\mathbb{Z}\times\mathbb{Z}\ ; x+y \text{ est impair} \}.
$$

En effet, les trajectoires sont obtenues en effectuant la somme d'un certain nombre `T` de "pas" $p_1, ..., p_T \in \mathbb{Z}\times\mathbb{Z}$, choisis aléatoirement parmi $(2,1), (-1,0),(0,1)$ et $(0,-1)$, qui sont tous dans $I$. Or on montre facilement que la somme de `T` éléments de $I$ est dans $P$ si `T` est pair, et dans $I$ si `T` est impair. Dans notre exemple, `T=100` est pair, c'est pourquoi les trajectoires arrivent toutes dans $P$ et aucune dans $I$.

### Un lissage gaussien $\triangleright$

Si l'on veut avoir une idée de la densité, il faut utiliser une méthode `kde` (=kernel-density estimate). L'idée est très simple, on place une petite gaussienne 2d en chaque donnée, et l'on somme le tout.

```python
from scipy.stats import kde

fig,ax=plt.subplots()

kernel = kde.gaussian_kde(arrivals.T)

x=np.linspace(gauche, droite,20)
XX,YY=np.meshgrid(x,x)
z=np.vstack([XX.reshape(-1),YY.reshape(-1)])
ZZ=kernel(z).reshape(XX.shape)

ax.imshow(ZZ,cmap="jet",interpolation="bilinear",origin="lower",extent=[gauche,droite,gauche,droite]);
```

Lorsqu'on "lisse" cet histogramme avec une méthode de noyaux, on obtient ce qui ressemble à la densité d'une gaussienne. Ça n'a rien d'étonnant, puisque le point d'arrivée d'une trajectoire est une somme de v.a indépendantes et de même loi (les "pas"), et donc en vertu du théorème central limite, ce point d'arrivée se rapproche d'une loi normale lorsque `T` tend vers l'infini.

### Et la vraie gaussienne? $\triangleright$

On peut estimer l'espérance et la covariance de la vraie gaussienne avec les fonctions `np.mean()` et `np.cov()`, mais attention, `np.cov()` attend en argument un tableau dont chaque colonne représente une observation, il faut donc transposer `arrivals`. Par excès de zèle, on peut aussi les calculer de manière exacte avec une feuille et un crayon ; on obtient alors
$$
\text{mean} =
\begin{pmatrix}
\frac{1}{4}T \\ \frac{1}{4} T
\end{pmatrix}
\quad\text{et}\quad
\text{cov} =
\begin{pmatrix}
\frac{19}{16}T & \frac{1}{2}T \\
\frac{1}{2}T & \frac{11}{16}T
\end{pmatrix},$$
où $T$ est le nombre de pas de la trajectoire. Afin de vérifier ce résultat, on superpose les contours de la densité de cette gaussienne à l'histogramme lissé obtenu précédemment.
Toujours par excès de zèle, on superpose également les histogrammes et les densités des lois marginales.

```python
from scipy.stats import kde

kernel = kde.gaussian_kde(arrivals.T)

x=np.linspace(gauche, droite,50)
xx,yy=np.meshgrid(x,x)
XY=np.stack([xx.reshape(-1),yy.reshape(-1)])
Z_kde=kernel(XY).reshape(xx.shape);
```

```python
T = 100

mean=np.mean(arrivals, axis=0)
print("mean empirique:\n", mean)
cov=np.cov(arrivals.T)
print("cov empirique:\n", cov)

mean=np.array([1/4*T, 1/4*T])
print("mean théorique:\n", mean)
cov=np.array([[19/16*T,1/2*T],[1/2*T,11/16*T]])
print("cov théorique:\n", cov)

fig=plt.figure(figsize=(6,6))
bins = np.linspace(gauche,droite+1,38)

# Histogramme et densité 2D
ax_xy = plt.subplot2grid((3, 3), (1, 0), colspan=2,rowspan=2)
ax_xy.set_aspect("equal")
ax_xy.imshow(Z_kde,interpolation="bilinear",
             origin="lower",extent=[gauche,droite,gauche,droite],
             cmap="jet",vmin=0,vmax=0.002,alpha = 0.8)

xy=np.stack([xx,yy],axis=2)
Z_density=stats.multivariate_normal.pdf(xy,mean=mean,cov=cov)
ax_xy.contour(xx,yy,Z_density,
              origin="lower",extent=[gauche,droite,gauche,droite],
              cmap="jet",vmin=0,vmax=0.002);

# Histogramme et densité des x
ax_x = plt.subplot2grid((3, 3), (0, 0), colspan=2)
ax_x.hist(arrivals[:,0],bins=bins,density=True,edgecolor="k")
ax_x.plot(bins,stats.norm.pdf(bins,mean[0],np.sqrt(cov[0,0])))

ax_x.set_xticks([])
ax_x.set_yticks([])

# Histogramme et densité des y
ax_y = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax_y.hist(arrivals[:,1],bins=bins,density=True,
          orientation='horizontal',edgecolor="k")
ax_y.plot(stats.norm.pdf(bins,mean[1],np.sqrt(cov[1,1])),bins)

ax_y.set_xticks([])
ax_y.set_yticks([]);
```

## Résumé des options de plt.hist()

Les arguments importants de `plt.hist()`:

L'argument obligatoire:
* `x:` les données

Argument sur le traitement des données:
* `bins`: découpage en sous-intervalles. Soit un nombre, soit une liste.  
* `density=True`: la hauteur des bâtons est normalisée pour que cela ressemble à une densité.
* `range=[gauche,droite]`: les données sont limitées à l'intervalle donné. Attention, l'option `density=True` donne dans ce cas un résultat bizarre. Il faut donc utiliser l'option suivante.
* `weights`: pondère les données. Pour superposer avec la densité `weights=np.ones_like(X)/step/len(X)` où `step` est la largeur de chaque sous-intervalle.

Options de présentation:
* `orientation:` `"horizontal"` ou  `"vertical"`
* `rwidth`: largeur des bâtons. Ex: `rwidth=0.5`, ils occupent la moitié des sous-intervalles.
* `edgecolor`: couleur du pourtour des bâtons. `"k"` pour black, `"red"` pour red, etc.
* `color`: couleur des bâtons.

```python
X=np.random.normal(size=100)
plt.hist(X,bins=10,color="red", density=True,rwidth=0.8,orientation="horizontal",edgecolor="k");
```
