## Reading and writing images

### images are arrays

Color images are 3d-arrays. Usualy, coefficient are unsign integer on 8 bits, so they range from 0 to 255.

```python
img_origin = imageio.imread("assets_signal/babouin/babouin_moyen.jpg")
img_origin.shape,img_origin.dtype
```

```python
img=img_origin.copy()

"""3 chanels : R,G,B = Red,Green,Blue.
They mix up to produce all possible colors"""
red = img[:,:,0]
green = img[:,:,1]
blue = img[:,:,2]

print(red[0:20,0:20])
```

```python
"we modify the image"
img[100:120,0:200,:]=0

fig,axs=plt.subplots(2,2)

""" if the arg of imshow is a 3d-array it mixes the colors """
axs[0,0].imshow(img)
axs[0,0].set_title("mix of colors")

""" if the arg of imshow is a 2d-array it uses a color map (here the gray-one) """
axs[0,1].imshow(red,cmap="gray")
axs[0,1].set_title("red")

axs[1,0].imshow(green,cmap="gray")
axs[1,0].set_title("green")

axs[1,1].imshow(blue,cmap="gray")
axs[1,1].set_title("blue")

fig.tight_layout()
```

### Black and white images

In the current language we speak of "black and white" images but actually the good name is "grey-level images".  To produce such an images, we just have two choices:

* produce a 3d-array, with the three channels equal
* produce a 2d-array, which we also consider as an image with 1channel.

Of course, the second method is more economic.


To change che baboin image into gray-level, we can just do the mean of the three channels.

```python
img=img_origin.copy()
gray_img = 0.333 * img[:,:,0] + 0.333 * img[:,:,1] + 0.333 * img[:,:,2]
plt.imshow(gray_img,cmap="gray");
```
