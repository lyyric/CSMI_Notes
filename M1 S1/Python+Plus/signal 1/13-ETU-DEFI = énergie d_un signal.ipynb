# Energie d'un signal

On considère $f$ un signal dérivable, périodique, sur $[0,T]$.


Nous définissons l'énergie de $f$ par:
$$
E(f)=\frac 1 T \int_0^T |f'(t)|^2\, dt
$$
(le prime désigne la dérivée en $t$)





***Remarque:*** Dans d'autres livres, l'énergie est définie par $\int_0^T |f(t)|^2\, dt$ (pas de dérivation). Mais cela dépend de la signification physique de $f$.

Par exemple, si $f$ représente une pression, il est plus naturelle de définir l'énergie avec une dérivation: Avec une pression constante, même grande, on ne peut pas produire d'électricité.   



## Question A

Quelle est l'énergie d'une vaguelette élementaire exponentielle $e_n$ ?  

### → ♡♡




$$
E(e_n)=\frac 1 T \int_0^T |e_n'(t)|^2\, dt  
$$



\begin{align}
&=\frac 1 T \int_0^T |\color{red}{\square \square \square} e_n(t)|^2\, dt \\
&=\color{red}{\square \square \square} \int_0^T 1 \, dt \\
&= \color{red}{\square \square \square}
\end{align}

A la fin on obtient une expression sans aucun signe intégral.


## Question B

 Comment peut-on exprimer l'énergie de $f$ en fonction de ses coefficients de fourier exponentiels.



### → ♡♡♡



Pour calculer l'énergie, on a utiliser le fait que la transformation de Fourier est une <font color="red"> □ □ □ </font>, ce qui donne:
$$
\frac 1 T \int_0^T |f'|^2 = \sum_n |\alpha_n(f')|^2
$$
où $\alpha_n(f')$ désigne ne $n$-ième coefficient de Fourier de $f'$.


Par Définition:
$$
\alpha_n(f') = \frac 1 T \int_0^T f' \color{red}{\square \square \square}
$$


En effectuant une <font color="red"> □ □ □ </font> on trouve:
$$
\alpha_n(f') =    \frac {-1} T \int_0^T f \bar e'_n
$$
$$
= \frac {-1} T \int_0^T \color{red}{\square \square \square} f  \bar e_n =\color{red}{\square \square \square} \alpha_n(f)
$$
Conclusion:
$$
\frac 1 T \int_0^T |f'|^2 = \sum_n \color{red}{\square \square \square} |\alpha_n(f)|^2
$$


Remarque: On confirme le résultat de la question A car, quand $f=e_m$, on a $\alpha_n(f) = 1_{n=m}$
