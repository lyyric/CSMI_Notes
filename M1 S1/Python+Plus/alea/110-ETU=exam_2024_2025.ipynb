## Vecteur en grande dimension


La fonction ci-dessous tire un vecteur aléatoire gaussien avec un écart-type bien précis:

Aide: `np.random.randn(n)` renvoie simplement `n` gaussiennes centrées réduites.



```python
im
def tire_vector(n):
    return np.random.randn(n)/np.sqrt(n)
```

## Régles du jeu





Deux joueurs J1 et J2 se font des tirs de handball. A chaque tour, chaque joueur tire 1 fois et l'autre se met dans les cages pour faire gardien de but. Si le tireur marque un but, il ajoute 1 point à son score.


A la fin de chaque tour on compte les points (et c'est équitable puisque les deux tireurs ont tiré le même nombre de fois).


Le match se joue en $n$ points gagnants,  mais il faut au moins 2 points d'écarts pour gagner. Sinon on continue jusqu'à qu'il y ait 2 points d'écarts.


Exemple: avec $n=8$.  

    J1:7, J2:1 => le jeu n'est pas fini
    J1:8, J2:1 => J1 gagne
    J1:1, J2:8 => J2 gagne
    J1:8, J2:6 => J1 gagne
    J1:8, J2:7 => le jeu n'est pas fini
    J1:9, J2:7 => J1 gagne
    J1:45, J2:44 => Le jeu n'est pas fini
    J1:46, J2:44 => J1 gagne



On suppose qu'à chaque tir J1 marque un but avec la probabilité 0.5. Alors que J2 marque un but avec la probabilité $p\in ]0,1[$.
